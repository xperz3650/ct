```python

dict1 = {}
chuoi_nhap1 = input("Nhập dictionary 1 (định dạng key:value, cách nhau bởi dấu phẩy, VD: a:1, b:2): ")
if chuoi_nhap1:
    cac_cap = chuoi_nhap1.split(',')
    for cap in cac_cap:
        k, v = cap.split(':')
        dict1[k.strip()] = int(v.strip())

dict2 = {}
chuoi_nhap2 = input("Nhập dictionary 2 (định dạng key:value, cách nhau bởi dấu phẩy, VD: b:3, c:4): ")
if chuoi_nhap2:
    cac_cap = chuoi_nhap2.split(',')
    for cap in cac_cap:
        k, v = cap.split(':')
        dict2[k.strip()] = int(v.strip())

dict_moi = {}
for k in dict1:
    dict_moi[k] = dict1[k]

for k in dict2:
    if k in dict_moi:
        dict_moi[k] += dict2[k]
    else:
        dict_moi[k] = dict2[k]

danh_sach_key = []
for k in dict_moi:
    danh_sach_key.append(k)

n = len(danh_sach_key)
for i in range(n):
    for j in range(0, n - i - 1):
        if danh_sach_key[j] > danh_sach_key[j + 1]:
            tam = danh_sach_key[j]
            danh_sach_key[j] = danh_sach_key[j + 1]
            danh_sach_key[j + 1] = tam

print("\nDictionary mới theo thứ tự từ điển tăng dần của key:")
for k in danh_sach_key:
    print(f"'{k}': {dict_moi[k]}")
```

```python
chuoi_nhap_list = input("Nhập danh sách các số nguyên (cách nhau bởi khoảng trắng): ")
danh_sach_chuoi = chuoi_nhap_list.split()
danh_sach_so = []
for s in danh_sach_chuoi:
    danh_sach_so.append(int(s))

k_str = input("Nhập số nguyên dương k: ")
k = int(k_str)

ket_qua = []
n = len(danh_sach_so)

for i in range(n):
    for j in range(i + 1, n):
        if danh_sach_so[i] + danh_sach_so[j] == k:
            cap_so = (danh_sach_so[i], danh_sach_so[j])
            
            da_ton_tai = False
            for x in ket_qua:
                if x == cap_so:
                    da_ton_tai = True
                    break

            if not da_ton_tai:
                ket_qua.append(cap_so)

so_luong_ket_qua = len(ket_qua)
for i in range(so_luong_ket_qua):
    for j in range(0, so_luong_ket_qua - i - 1):
        if ket_qua[j][0] > ket_qua[j + 1][0]:
            tam = ket_qua[j]
            ket_qua[j] = ket_qua[j + 1]
            ket_qua[j + 1] = tam

print("Danh sách các cặp số:", ket_qua)
```

```python
chuoi_nhap = input("Nhập danh sách các số nguyên (cách nhau bởi khoảng trắng): ")
danh_sach_chuoi = chuoi_nhap.split()

danh_sach_so = []
for s in danh_sach_chuoi:
    danh_sach_so.append(int(s))

# Dùng dictionary để đếm số lần xuất hiện
tu_dien_dem = {}
for so in danh_sach_so:
    if so in tu_dien_dem:
        tu_dien_dem[so] += 1
    else:
        tu_dien_dem[so] = 1

so_xuat_hien_nhieu_nhat = None
so_lan_max = 0

for so in tu_dien_dem:
    so_lan = tu_dien_dem[so]

    if so_lan > so_lan_max:
        so_lan_max = so_lan
        so_xuat_hien_nhieu_nhat = so
    elif so_lan == so_lan_max:
        if so < so_xuat_hien_nhieu_nhat:
            so_xuat_hien_nhieu_nhat = so

print("Số xuất hiện nhiều nhất (ưu tiên số nhỏ nhất):", so_xuat_hien_nhieu_nhat)
```

```python
chuoi1 = input("Nhập danh sách thứ nhất (các số cách nhau bởi khoảng trắng): ")
list_chuoi1 = chuoi1.split()
list_so1 = []
for s in list_chuoi1:
    list_so1.append(int(s))

chuoi2 = input("Nhập danh sách thứ hai (các số cách nhau bởi khoảng trắng): ")
list_chuoi2 = chuoi2.split()
list_so2 = []
for s in list_chuoi2:
    list_so2.append(int(s))

list_chung = []

for so1 in list_so1:
    co_trong_list2 = False
    for so2 in list_so2:
        if so1 == so2:
            co_trong_list2 = True
            break
    if co_trong_list2:
        da_co_trong_list_chung = False
        for chung in list_chung:
            if so1 == chung:
                da_co_trong_list_chung = True
                break

        if not da_co_trong_list_chung:
            list_chung.append(so1)

print("Danh sách các phần tử chung:", list_chung)
```
