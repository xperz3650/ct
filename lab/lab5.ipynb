```python
def prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
if prime(n):
    print("YES")
else:
    print("NO")
```

```python
def perfect_number(n):
    if n < 2:
        return False
    sum_of_divisors = 1
    for i in range(2, n):
        if n % i == 0:
            sum_of_divisors += i
    return sum_of_divisors == n
n = int(input())
if perfect_number(n):
    print("YES")
else:
    print("NO")
```

```python
def tim_vi_tri_dau_tien(list, k):
    for i in range(len(list)):
        if list[i] == k:
            return i  
    return -1 

list = list(map(int, input().split()))
k = int(input())

vi_tri = tim_vi_tri_dau_tien(list, k)
print("Vị trí của phần tử đầu tiên có giá trị", k, "là:", vi_tri)
```

```python
def factorial(n):
    if n < 0:
        return None
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
n = int(input())
print(factorial(n))
```

```python
def calculator(a, b, operator):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
        if b != 0:
            return a / b
        else:
            return "Error: Division by zero"
    else:
        return "Error: Invalid operator"
a = float(input())
b = float(input())
operator = input()
result = calculator(a, b, operator)
print("Kết quả:", result)
```

```python
def hamming_distance(str1, str2):
    if len(str1) != len(str2):
        return None
    distance = 0
    for c1, c2 in zip(str1, str2):
        if c1 != c2:
            distance += 1
    return distance
str1 = input()
str2 = input()
print(hamming_distance(str1, str2))
```

```python
def total_of_digits(n):
    total = 0
    while n > 0:
        total += n % 10
        n //= 10
    return total
n = int(input())
print(total_of_digits(n))
```

```python
def divide_with_remainder(dividend, divisor):
    r = dividend % divisor
    q = dividend // divisor
    if dividend == divisor * q + r and 0 <= r < divisor:
        return q, r
    
dividend = int(input())
divisor = int(input())
q, r = divide_with_remainder(dividend, divisor)
print("Quotient:", q)
print("Remainder:", r)
```

```python
def is_right_triangle(a, b, c):
    sides = sorted([a, b, c])
    return sides[0]**2 + sides[1]**2 == sides[2]**2
a = float(input())
b = float(input())
c = float(input())
if is_right_triangle(a, b, c):
    print("True")
else:
    print("False")
```
