```python
chuoi_nhap = input("Nhập các chuỗi (tách bởi dấu cách): ")
danh_sach_chuoi = chuoi_nhap.split()

tu_dien_vi_tri = {}

for i in range(len(danh_sach_chuoi)):
    chuoi = danh_sach_chuoi[i]
    
    if chuoi not in tu_dien_vi_tri:
        tu_dien_vi_tri[chuoi] = []
        
    tu_dien_vi_tri[chuoi].append(i)

for chuoi in tu_dien_vi_tri:
    tu_dien_vi_tri[chuoi] = tuple(tu_dien_vi_tri[chuoi])

print("Dictionary kết quả:")
print(tu_dien_vi_tri)
```

```python
def xoa_so_nho_hon_x(mang, x):
    mang_ket_qua = []
    
    for i in range(len(mang)):
        # Chỉ giữ lại những phần tử >= x
        if mang[i] >= x:
            mang_ket_qua.append(mang[i])
            
    return mang_ket_qua

mang_ban_dau = [1, 5, 8, 3, 2, 9]
print(xoa_so_nho_hon_x(mang_ban_dau, 4))
```

```python
def day_khong_ve_dau(mang):
    mang_ket_qua = []
    
    for i in range(len(mang)):
        if mang[i] == 0:
            mang_ket_qua.append(0)

    for i in range(len(mang)):
        if mang[i] != 0:
            mang_ket_qua.append(mang[i])
            
    return mang_ket_qua


mang_thu = [0, 2, 3, 2, 4, 0, 0, 0, 9]
print(day_khong_ve_dau(mang_thu))
```

```python
chuoi_nhap = input("Nhập các số trong danh sách: ").split()
danh_sach = []
for s in chuoi_nhap:
    danh_sach.append(int(s))

tong_cuc_dai = 0

for i in range(1, len(danh_sach) - 1):
    if danh_sach[i] > danh_sach[i - 1] and danh_sach[i] > danh_sach[i + 1]:
        tong_cuc_dai += danh_sach[i]

print(f"Tổng các phần tử cực đại địa phương: {tong_cuc_dai}")
```
