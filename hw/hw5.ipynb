```python
def is_armstrong(x: int) -> bool:
    if x < 0:
        return False
    str_x = str(x)
    num_digits = len(str_x)
    total = sum(int(digit) ** num_digits for digit in str_x)
    return total == x
n = int(input())
print(is_armstrong(n))
```

```python
def is_palindrome(x: int) -> bool:
    if x < 0:
        return False
        
    original_x = x
    reversed_x = 0
    
    while x > 0:
        last_digit = x % 10
        reversed_x = reversed_x * 10 + last_digit
        x = x // 10
        
    return original_x == reversed_x

n = int(input())
print(is_palindrome(n))
```

```python
def is_magic_number(x: int) -> bool:
    if x <= 0:
        return False
    while x > 9:
        sum_digits = 0

        while x > 0:
            last_digit = x % 10
            sum_digits += last_digit
            x = x // 10
        x = sum_digits
        
    return x == 1
n = int(input())
print(is_magic_number(n))
```

```python
def point_in_quadrilateral(quad: list[tuple[float, float]], p: tuple[float, float]) -> bool:
    x, y = p
    n = len(quad)
    inside = False

    for i in range(n):
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % n]
        cross_product = (x - x1) * (y2 - y1) - (y - y1) * (x2 - x1)
        if abs(cross_product) < 1e-9:
            if min(x1, x2) <= x <= max(x1, x2) and min(y1, y2) <= y <= max(y1, y2):
                return True # Điểm nằm ngay trên biên
        if (y1 > y) != (y2 > y):
            x_intersect = (x2 - x1) * (y - y1) / (y2 - y1) + x1
            if x < x_intersect:
                inside = not inside

    return inside
quad_points = [(0,0), (4,0), (4,3), (0,3)]
point_to_check = (2, 1)
print(point_in_quadrilateral(quad_points, point_to_check))
```

```python
def count_non_zero_digits(numbers: list[int]) -> int:
    total_count = 0
    
    for num in numbers:
        str_num = str(abs(num))
        for digit in str_num:
            if digit != '0':
                total_count += 1
                
    return total_count
numbers = [int(x) for x in input().split()]
print(count_non_zero_digits(numbers))
```

```python
import math
def calculate_angle_abc(x1: float, y1: float, x2: float, y2: float, x3: float, y3: float) -> float:
    ba_x = x1 - x2
    ba_y = y1 - y2
    
    bc_x = x3 - x2
    bc_y = y3 - y2
    
    dot_product = ba_x * bc_x + ba_y * bc_y
    mag_ba = math.hypot(ba_x, ba_y)
    mag_bc = math.hypot(bc_x, bc_y)
    if mag_ba == 0 or mag_bc == 0:
        return 0.0

    cos_angle = max(-1.0, min(1.0, dot_product / (mag_ba * mag_bc)))
    angle_rad = math.acos(cos_angle)
    return math.degrees(angle_rad)

x1 = float(input())
y1 = float(input())
x2 = float(input())
y2 = float(input())
x3 = float(input())
y3 = float(input())
print(calculate_angle_abc(x1, y1, x2, y2, x3, y3))
```

```python
def smart_taxi_fare(car_type: str, start_hour: float, km: float, wait_minutes: int) -> float:
    OPENING_PRICE = 10000         
    OPENING_KM = 0.5              
    
    PRICE_PER_KM = 15000 if car_type == '4_seats' else 17000 
    
    WAIT_FEE_PER_MIN = 800        
    FREE_WAIT_MINS = 5            
    
    NIGHT_SURCHARGE_RATE = 0.1    
    
    DISCOUNT_RATE = 0.05         
    DISCOUNT_KM_THRESHOLD = 30   
    total_fare = 0.0

    if km <= OPENING_KM:
        total_fare += OPENING_PRICE
    else:
        total_fare += OPENING_PRICE + (km - OPENING_KM) * PRICE_PER_KM

    if wait_minutes > FREE_WAIT_MINS:
        total_fare += (wait_minutes - FREE_WAIT_MINS) * WAIT_FEE_PER_MIN

    if start_hour >= 22 or start_hour <= 5:
        total_fare += total_fare * NIGHT_SURCHARGE_RATE

    if km > DISCOUNT_KM_THRESHOLD:
        total_fare -= total_fare * DISCOUNT_RATE

    return total_fare

car_type = input("Nhập loại xe (4_seats hoặc 7_seats): ")
start_hour = float(input("Nhập giờ bắt đầu (0-23): "))
km = float(input("Nhập số km đã đi: "))
wait_minutes = int(input("Nhập số phút chờ: "))
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

```python
def sum_1_to_n(n: int) -> int:
    if n <= 0:
        return 0
    total_sum = 0
    for x in range(1, n + 1):
        str_x = str(x)
        if len(set(str_x)) == len(str_x):
            total_sum += x
            
    return total_sum

n = int(input())
print(sum_1_to_n(n))
```
