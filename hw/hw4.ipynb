```python
n = int(input("Nhập một số nguyên dương n: "))
k = 0
while 2**k < n:
    k += 1
if 2**k == n:
    print(f"{n} là một số nguyên dương có dạng 2^k với k = {k}.")
else:
    print(f"{n} không phải là một số nguyên dương có dạng 2^k.")
```

```python
n = int(input("Nhập một số nguyên dương n: "))
k = 1
Sk = (k * (k + 1)) // 2
while Sk < n:
    k += 1
    Sk = (k * (k + 1)) // 2
    if Sk >= n:
        break
    else:
        continue
print(f"Số tự nhiên k lớn nhất sao cho Sk < {n} là: {k}")
```

```python
n = int(input("Nhập một số nguyên dương n: "))
k = 1
h = 1/k + 1/(k+1)
while h <= n:
    h += 1 / (k+1)
    if h > n:
        break
    else:
        continue
print(f"Số tự nhiên k lớn nhất sao cho h > {n} là: {k}")
```

```python

max = None
min = None
print("Nhập các số nguyên (nhập -1 để dừng):")
while True:
    so = int(input())
    if so == -1:
        break
    if max is None or so > max:
        max = so
    if min is None or so < min:
        min = so
if max is not None:
    print(f"Số lớn nhất: {max}")
    print(f"Số nhỏ nhất: {min}")        
else:
    print("Bạn chưa nhập số nào hợp lệ trước khi dừng.")
```

```python
chuoi = input("Nhập chuỗi: ")
do_dai = 0
for ky_tu in chuoi:
    do_dai += 1
so_tu = 0
dang_o_trong_tu = False
for i in range(do_dai):
    ky_tu_hien_tai = chuoi[i]
    if ky_tu_hien_tai != ' ':
        if dang_o_trong_tu == False:
            so_tu += 1
            dang_o_trong_tu = True
    else:
        dang_o_trong_tu = False
print(f"Số từ trong chuỗi là: {so_tu}")
```
